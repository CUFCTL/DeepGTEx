```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
doubles = np.genfromtxt("hedgehog_two_gene_accuracy.txt",  delimiter='\t',dtype=str)
singles = np.genfromtxt("hedgehog_one_gene_accuracy.txt",  delimiter='\t',dtype=str)
b = doubles[:,1]
b = b.astype(np.float32)
np.max(b)
```

```python
doubles[:,1] = b
```

```python
doubles = doubles[np.argsort(doubles[:, 1])]
doubles_top = doubles[len(doubles)-10:]
doubles_bottom = doubles[:10]
```

```python
doubles_top
```

```python
labels = np.array(doubles_bottom[:,0])
labels = np.append(labels,(doubles_top[:,0]))

values = np.array(doubles_bottom[:,1])
values = np.append(values,(doubles_top[:,1]))
```

```python
values
```

```python
labels
```

```python
values = values.astype(np.float32)
```

```python
fig = plt.figure(figsize=(20,10))
x = np.arange(len(labels))
plt.bar(x,values.tolist())
plt.xlabel('Gene Pairs')
plt.title("Pair Accuracies Top/Bottom 10")
plt.xticks(x,labels)
plt.xticks(rotation=90)
plt.ylabel("predictions")
plt.show()
```
