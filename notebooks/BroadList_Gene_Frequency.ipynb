```python
#/usr/bin/python

import numpy as np
import os
import matplotlib.pyplot as plt

hallmarks = []
gtex_string_data = []
ensembl_map = []
hallmark_dict = dict()
genes_dict = dict()
```

```python
print('loading hallmark gene list...')
with open('../BROAD_LISTS/h.all.v6.0.symbols.gmt') as f:
	content = f.readlines()

for c in content:
	c = c.replace('\n', '')
	hallmarks.append(np.array(c.split('\t')))
```

```python
for i in range(len(hallmarks)):
    hallmark_dict[hallmarks[i][0]] = hallmarks[i][2:len(hallmarks[i])]
```

```python
hallmark_dict['HALLMARK_PANCREAS_BETA_CELLS']
```

```python
count = 0
for i in hallmark_dict.keys():
    print(i)
    print(len(hallmark_dict[i]))
    count = count + len(hallmark_dict[i])
```

```python
count
```

```python
for i in hallmark_dict.keys():
    gene_array = hallmark_dict[i]
    for j in range(len(hallmark_dict[i])):
        if gene_array[j] in genes_dict:
             genes_dict[gene_array[j]] += 1
        else:
            genes_dict[gene_array[j]] = 1
```

```python
count = 0
for i in genes_dict.keys():
    count = count + genes_dict[i]
```

```python
unique_genes = { k:v for k, v in genes_dict.items() if v > 4}
```
