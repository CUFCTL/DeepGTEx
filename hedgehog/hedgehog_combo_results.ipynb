```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.genfromtxt("results.txt",  delimiter='\t',dtype=str)
```

```python
#Max/Min/Avg/Var/Std
maxs = list()
mins = list()
avgs = list()
var = list()
stds = list()

for i in range(len(results)):
    maxs.append(float(results[i][0]))
    mins.append(float(results[i][1]))
    avgs.append(float(results[i][2]))
    var.append(float(results[i][3]))
    stds.append(float(results[i][4]))
    
```

```python
fig = plt.figure(figsize=(20,10))
x = np.arange(len(maxs))
plt.bar(x,maxs)
plt.xlabel('Combo Amounts')
plt.title("Max Values across 36 gene combos")
plt.ylabel("Prediction Level")
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
x = np.arange(len(maxs))
plt.bar(x,mins)
plt.xlabel('Combo Amounts')
plt.title("Min Values across 36 gene combos")
plt.ylabel("Prediction Level")
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
x = np.arange(len(avgs))
plt.bar(x,avgs)
plt.xlabel('Combo Amounts')
plt.title("Avg Values across 36 gene combos")
plt.ylabel("Prediction Level")
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
x = np.arange(len(avgs))
plt.bar(x,var)
plt.xlabel('Combo Amounts')
plt.title("Var Values across 36 gene combos")
plt.ylabel("Prediction Level")
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
x = np.arange(len(avgs))
plt.bar(x,stds)
plt.xlabel('Combo Amounts')
plt.title("Std Values across 36 gene combos")
plt.ylabel("Prediction Level")
plt.show()
```
